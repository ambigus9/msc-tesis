```python
def graficos_barras_v2(df_fusion, save_name):
    df_temp = df_fusion.copy()
    # GRAFICO DE BARRAS ESTABLE CON PORCENTAJE
    import plotly.express as px
    fig = px.bar(df_temp.dropna(), x='arquitectura', y='training_time', barmode='group', color='dataset',
        height=500, text='iteracion')

    fig.update_layout(
        title='Training time',
        yaxis_title='Time (min)',
        xaxis_title='Architecture', #)
        xaxis={'categoryorder':'array', 'categoryarray':['ucmerced','WHU-RS19','AID',
                                                        'NWPU-RESISC45']})
    #fig.update_traces(texttemplate='%{text:.2s}%', textposition='outside', textangle=90, textfont_size=6)
    fig.update_layout(uniformtext_minsize=12, uniformtext_mode='show')
    #fig.update_layout(yaxis=dict(range=[0,850e3]))
    #fig.write_html('gb_cuentaContable_cruzo.html')
    fig.write_image(save_name)
    return fig.show()
 
#df_time_abs
graficos_barras_v2(df_abs_time, 'training_time_satellital.pdf')
```
